```python
import gymnasium as gym
from gymnasium import spaces
import numpy as np
import pickle
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import pylab as plt
```

## Helper Functions

```python
def running_mean(x, N):
    cumsum = np.cumsum(np.insert(x, 0, 0)) 
    return (cumsum[N:] - cumsum[:-N]) / float(N)    
```

# PPO parameter dependency

## clip (epsilon)

```python
# You can load the files with pickle.load 
#  with open(filename, 'rb') as f:....
# consider using a for-loop over the different eps values
for eps in [0.01, 0.1, 0.2, 0.5, 0.75]:
    filename = f'./results/PPO_LunarLander-v3-eps{eps}-stat.pkl'
    with open(filename, 'rb') as f:
        data = pickle.load(f)
    plt.figure()
    plt.plot(running_mean(data['rewards'], 20))
    plt.xlabel('episodes')
    plt.ylabel('avg rewards')
    plt.title(f'eps_clip = {eps}')
```
